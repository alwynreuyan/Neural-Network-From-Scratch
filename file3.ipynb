```python
import cv2
```

```python
capture = cv2.VideoCapture('C:/Users/Alwyn/OneDrive/Desktop/Neural-Network-From-Scratch/LAB-4.mp4');

while(True):
    
    ret, frame = capture.read()
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('Video Player', gray)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
capture.release()
cv2.destroyAllWindows()
```
